```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ecuacion diferencial a resolver x''  + (g/l)x = 0
g = 9.81
l = 1.0
w2 = g/l 
```

```python
#Condiciones iniciales
x0 = 1
v0 = 0
```

```python
#Funciones para el método de runge kutta
def f(x):
    return -w2*np.sin(x)    
```

```python
def g(v):
    return v
```

```python
#programa principal
to = 0.0
tf = 10.0
n = 500
h = (tf - to)/n
time = np.zeros(n)
theta = np.zeros(n)
for k in range(1,n):
    t = to + (k-1)*h
    k1 = h*g(v0)
    l1 = h*f(x0)
    k2 = h*g(v0 + 0.5*l1)
    l2 = h*f(x0 + 0.5*k1)
    k3 = h*g(v0 + 0.5*l2)
    l3 = h*f(x0 + 0.5*k2)
    k4 = h*g(v0 + l3)
    l4 = h*f(x0 + k3)
    v1 = v0 + (l1 + 2*l2 + 2*l3 + l4)/6
    x1 = x0 + (k1 + 2*k2 + 2*k3 + k4)/6
    time[k] = t + h
    theta[k] = x1
    v0 = v1
    x0 = x1
```

```python
#graficamos theta vs t
plt.plot(time,theta,label = r'Solución $\theta(t)$')
plt.xlabel('t')
plt.ylabel(r'$\theta$')
plt.title(r'Grafica solución a $\frac{d²}{dt²}\theta + \frac{g}{L}\sin(\theta) = 0$')
plt.legend()
plt.savefig("solucion_pendulo.png")
plt.show()
```
